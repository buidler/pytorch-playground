### 向量

向量是一个一维的tensor，类似于一维数组

```python
vector = torch.tensor([7,8])
print(vector)
```

#### vector的维度（和数组类似）

外层有几个[，就有几个维度

```python
vector.ndim
```

#### 数据的形状

```python
vector.shape
```
