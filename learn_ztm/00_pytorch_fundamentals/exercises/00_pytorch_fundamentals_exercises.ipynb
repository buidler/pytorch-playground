### 2. Create a random tensor with shape `(7, 7)`.


````python
# Import torch
import torch


# Create random tensor
random_tensor = torch.randn(7, 7)
print(random_tensor)
print(random_tensor.shape)
print(random_tensor.ndim)
````

### 3. Perform a matrix multiplication on the tensor from 2 with another random tensor with shape `(1, 7)` (hint: you may have to transpose the second tensor).

````python
# Create another random tensor
another_random_tensor = torch.randn(1, 7)
another_random_tensor = torch.transpose(another_random_tensor, 0, 1)
print(another_random_tensor)
# Perform matrix multiplication 
matrix_multiplication = torch.matmul(random_tensor, another_random_tensor)
print(matrix_multiplication)
````

### 8. Find the maximum and minimum values of the output of 7.

````python
# Find max
print(torch.matmul(tensor_random_g_1, torch.transpose(tensor_random_g_2, 0, 1)).max())
# Find min
print(torch.matmul(tensor_random_g_1, torch.transpose(tensor_random_g_2, 0, 1)).min())
````

### 9. Find the maximum and minimum index values of the output of 7.


### 10. Make a random tensor with shape `(1, 1, 1, 10)` and then create a new tensor with all the `1` dimensions removed to be left with a tensor of shape `(10)`. Set the seed to `7` when you create it and print out the first tensor and it's shape as well as the second tensor and it's shape.

创建一个形状为 (1, 1, 1, 10) 的随机张量，然后创建一个新张量 —— 移除所有尺寸为 1 的维度，最终得到形状为 (10) 的张量。创建随机张量时将随机种子设置为 7，并打印第一个张量及其形状，以及第二个张量及其形状。

The output should look like:

```
tensor([[[[0.5349, 0.1988, 0.6592, 0.6569, 0.2328, 0.4251, 0.2071, 0.6297,
           0.3653, 0.8513]]]]) torch.Size([1, 1, 1, 10])
tensor([0.5349, 0.1988, 0.6592, 0.6569, 0.2328, 0.4251, 0.2071, 0.6297, 0.3653,
        0.8513]) torch.Size([10])
```

````python
# Set seed
torch.cuda.manual_seed(7)
device = torch.device("cuda:0")

# Create random tensor
tensor_random_g_seed7 = torch.randn(1,1,1,10,device=device)

# Remove single dimensions
tensor_random_g_seed7_squ = tensor_random_g_seed7.squeeze()

# Print out tensors and their shapes
print(tensor_random_g_seed7)
print(tensor_random_g_seed7.shape)
print(tensor_random_g_seed7_squ)
print(tensor_random_g_seed7_squ.shape)
````
